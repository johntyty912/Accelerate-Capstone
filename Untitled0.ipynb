```
import pandas as pd
import numpy as np

import re
import os
import csv

from pykalman import KalmanFilter
import matplotlib.pyplot as plt
```

```
path = "drive/capstone/"
```

```
CCL = pd.read_csv(path+"datas/CCL.csv", sep="\t")
```

```
# Set up the filter
kf = KalmanFilter(transition_matrices = [1],
                  observation_matrices = [1],
                  initial_state_mean = CCL['CCL'].values[0],
                  initial_state_covariance = 1,
                  em_vars=["observation_covariance","transition_covariance"])
```

```
kf.em(CCL.CCL,n_iter=2)
```

```
state_means, state_covs = kf.filter(CCL.CCL)
```

```
plt.figure(figsize=(20,15))
plt.plot(CCL.Date, state_means[:,0])
plt.plot(CCL.Date, CCL.CCL)
plt.legend(['Kalman Filter', 'Actual'])
plt.xlabel('Time')
plt.ylabel('Index');
```
