```
import pandas as pd
import numpy as np
```

```
path="drive/capstone/"
```

```
hk = pd.read_csv(path+"datas/28hse_HongKong.csv")
```

```
kl = pd.read_csv(path+"datas/kowloon.csv", index_col=0)
```

```
nt = pd.read_csv(path+"datas/new_territories.csv", index_col=0)
```

```
nt.head()
```

```
nt["catfathername"].unique()
```

```
nt = nt[(~nt["catfathername"].isin(np.append(hk["catfathername"].unique(),kl["catfathername"].unique())))]
```

```
kl["catfathername"].unique()
```

```
hk["catfathername"].unique()
```

```
nt["Date"] = pd.to_datetime(nt.date, format="%Y-%m-%d", errors='ignore')
```

```
nt.set_index("Date", inplace=True)
```

```
nt = nt[nt["sq_price_value"] != 0.0]
```

```
len(nt.catfathername.unique())
```

```
#nt_price = pd.crosstab(nt.index.to_period("W"),nt.catfathername, values=nt["sq_price_value"], aggfunc='mean', margins=False).fillna(0.0)
nt_price = nt.groupby(pd.Grouper(freq='W'))["sq_price_value"].mean().to_frame().rename(index=str, columns={"sq_price_value":"nt_sq_price_value"})
```

```
for data in ["kl","hk"]:
  i=eval(data)
  i["Date"] = pd.to_datetime(i.date, format="%Y-%m-%d", errors='ignore')
  i.set_index("Date", inplace=True)
  i = i[i["sq_price_value"] != 0.0]
  i_price = i.groupby(pd.Grouper(freq='W'))["sq_price_value"].mean().to_frame().rename(index=str, columns={"sq_price_value":"{}_sq_price_value".format(data)})
  nt_price = pd.merge(nt_price,i_price,left_index=True,right_index=True)
#nt_price.to_csv(path+"datas/avg_house_price.csv")
```

```
nt_price.to_csv(path+"datas/avg_house_price.csv")
```
